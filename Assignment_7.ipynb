Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each
row.
iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and
each row

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_array = np.abs(array)
print("Element-wise Absolute Value:\n", abs_array)
flat_array = array.flatten()
flat_percentiles = np.percentile(flat_array, [25, 50, 75])
print("\nPercentiles (Flattened Array):", flat_percentiles)
col_percentiles = np.percentile(array, [25, 50, 75], axis=0)
print("\nPercentiles (Each Column):\n", col_percentiles)
row_percentiles = np.percentile(array, [25, 50, 75], axis=1)
print("\nPercentiles (Each Row):\n", row_percentiles)
flat_mean = np.mean(flat_array)
flat_median = np.median(flat_array)
flat_std = np.std(flat_array)
print("\nMean, Median, Std (Flattened):", flat_mean, flat_median, flat_std)
col_mean = np.mean(array, axis=0)
col_median = np.median(array, axis=0)
col_std = np.std(array, axis=0)
print("\nMean, Median, Std (Each Column):\n", col_mean, col_median, col_std)
row_mean = np.mean(array, axis=1)
row_median = np.median(array, axis=1)
row_std = np.std(array, axis=1)

print("\nMean, Median, Std (Each Row):\n", row_mean, row_median, row_std)
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
def swap_elements(lst, index1, index2):
    if index1 < 0 or index2 < 0 or index1 >= len(lst) or index2 >= len(lst):
        print("Invalid indices! Please enter valid positions.")
        return lst
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp

    return lst

my_list = [10, 20, 30, 40, 50]
index1, index2 = 1, 3
swapped_list = swap_elements(my_list, index1, index2)
print("List after swapping:", swapped_list)
```

Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.

```python
def swap_elements_in_set(s, elem1, elem2):
    if elem1 not in s or elem2 not in s:
        print("Error: One or both elements not found in the set!")
        return s

    lst = list(s)

    index1, index2 = lst.index(elem1), lst.index(elem2)

    lst[index1], lst[index2] = lst[index2], lst[index1]

    swapped_set = set(lst)

    return swapped_set

my_set = {10, 20, 30, 40, 50}

elem1, elem2 = 20, 40

swapped_set = swap_elements_in_set(my_set, elem1, elem2)

print("Set after swapping :", swapped_set)
```
