Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
i. Sum of all elements
ii. Sum of all elements row-wise
iii. Sum of all elements column-wise

```python
import numpy as np

gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
print(gfg)

gfgsum=np.sum(gfg)
print(gfgsum)

row_sum=np.sum(gfg,axis=1)
print(row_sum)

col_sum=np.sum(gfg,axis=0)
print(col_sum)
```

Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
i. Sorted array
ii. Indices of sorted array
iii. 4 smallest elements
iv. 5 largest elements

```python
array=np.array([10,52,62,16,16,54,453])
print(array)

sorted_array=np.sort(array)
print(sorted_array)

sorted_indices=np.argsort(array)
print(sorted_indices)

smallest_4=np.partition(array,4)[:4]
print(smallest_4)

largest_5=np.partition(array,-5)[-5:]
print(largest_5)
```

Q.2 (b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
i. Integer elements only
ii. Float elements only

```python
array=np.array([1.0,1.2,2.2,2.0,3.0,2.0])
print(array)

int_elements=array[array==array.astype(int)]
print(int_elements)

float_elements=array[array!=array.astype(int)]
print(float_elements)
```

Q.3 You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.
a) Generate your unique sales dataset:
 Take the sum of the ASCII values of the initials of your first and last
name. Call this value X. (If your initials are A B → ASCII sum = 65
+ 66 = 131 → sales = [131, 181, 231, 281, 331].)
 Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].

b) Compute your personalized tax rate as ((X % 5) + 5) / 100.
 Use broadcasting to apply this tax rate to each sales value.
c) Adjust sales based on discount:
 If sales < X+100, apply a 5% discount.
 If sales >= X+100, apply a 10% discount.
d) Expand sales data for multiple weeks:
 Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.
 Increase sales by 2% per week using element-wise broadcasting.

```python
first_name='BHAVYA'
last_name='KALAL'
ascii_sum=ord(first_name[0])+ord(last_name[0])
print(ascii_sum)

x=ascii_sum

sales=np.array([x,x+50,x+100,x+150,x+200])
print(sales)

tax_rate=((x%5)+5)/100
print(tax_rate)

sales_tax=sales*(tax_rate)
print(sales_tax)

total_sales=sales+sales_tax
print(total_sales)

discount_sales=np.where(total_sales<x+100,total_sales*0.95,total_sales*0.9)
print(discount_sales)

weekly_sales=np.tile(sales,(3,1))
print(weekly_sales)

weekly_growth=1.02**np.arange(3).reshape(3,1)
adjusted_sales=weekly_sales*weekly_growth
print(adjusted_sales)
```

Q.4 Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
 Y = x2
 Y = sin(x)
 Y = ex
 Y = log(|x| + 1)
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)

y1=x**2
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)

fig, axs=plt.subplots(2,2,figsize=(12,10))

axs[0,0].plot(x,y1)
axs[0,0].set_title('x^2')
axs[0,0].set_xlabel('x')
axs[0,0].set_ylabel('y')
axs[0,0].grid(True)

axs[0,1].plot(x,y2)
axs[0,1].set_title('sin(x)')
axs[0,1].set_xlabel('x')
axs[0,1].set_ylabel('y')
axs[0,1].grid(True)

axs[1,0].plot(x,y3)
axs[1,0].set_title('exp(x)')
axs[1,0].set_xlabel('x')
axs[1,0].set_ylabel('y')
axs[1,0].set_ylim(0,1000)
axs[1,0].grid(True)

axs[1,1].plot(x,y4)
axs[1,1].set_title('log(1+abs(x))')
axs[1,1].set_xlabel('x')
axs[1,1].set_ylabel('y')

axs[1,1].grid(True)

plt.tight_layout()
plt.show()

```
